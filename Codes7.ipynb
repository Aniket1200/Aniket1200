```python
import cv2
from google.colab.patches import cv2_imshow  # For displaying image in Colab

```

```python


# Define image (replace with your image path if needed)
img = cv2.imread('/content/drive/MyDrive/PU.jpg')  # Assuming you have an image named 'image.jpg'

# Define rectangle coordinates (top-left and bottom-right corners)
top_left = (50, 50)  # X, Y coordinates for top-left corner
bottom_right = (250, 300)  # X, Y coordinates for bottom-right corner

# Define rectangle color (Blue in BGR format)
color = (255, 0, 0)  # Blue color (B, G, R)

# Draw the rectangle with thickness of 2
cv2.rectangle(img, top_left, bottom_right, color, 2)

# Display the image
cv2_imshow(img)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

```

```python
# Define rectangle dimensions (adjust as needed)
width = 3000
height =3000

# Load image (replace 'path/to/your/image.jpg' with your actual image path)
image = mpimg.imread('/content/drive/MyDrive/PU.jpg')

# Create the main figure and plot area
fig, ax = plt.subplots()

# Create rectangle object with desired color and transparency (optional)
rectangle = plt.Rectangle((2, 2), width, height, color='black', alpha=0.4)
ax.add_patch(rectangle)

# Insert image within the rectangle (adjust x and y as needed to position)
ax.imshow(image, extent=(5, width, 5, height), aspect='auto')

# Set limits slightly higher than rectangle dimensions (optional)
ax.set_xlim(-width/50, width + width/50)
ax.set_ylim(-height/50, height + height/50)

# Remove axes (optional)
ax.axis('off')

plt.title('')
plt.show()
```
