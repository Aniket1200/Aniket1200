```python
import pandas as pd
```

```python
path="/content/drive/MyDrive/Alipurduar temp data.csv"
df = pd.read_csv(path)
df.head(253)
```

```python
# @title Moving average temp(y')

from matplotlib import pyplot as plt
df["Moving average temp(y')"].plot(kind='line', figsize=(8, 4), title="Moving average temp(y')")
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Unadjusted monthly average

from matplotlib import pyplot as plt
df['Unadjusted monthly average'].plot(kind='line', figsize=(8, 4), title='Unadjusted monthly average')
plt.gca().spines[['top', 'right']].set_visible(False)
```
